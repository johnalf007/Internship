```python
webpage=BeautifulSoup(page.content)
webpage
```

```python
house_title=[]
for i in webpage.find_all('span',class_="overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full"):
    house_title.append(i.text)
    
house_title
```

```python
location=[]
for i in webpage.find_all('div',class_="mt-0.5p overflow-hidden overflow-ellipsis whitespace-nowrap max-w-70 text-gray-light leading-4 po:mb-0 po:max-w-95"):
    location.append(i.text)
    
location
```

```python
area=[]
for i in webpage.find_all('span',class_="flex flex-wrap ml-0.5p tp:h-4p tp:overflow-hidden tp:overflow-x-scroll tp:w-full tp:whitespace-nowrap tp:block tp:-mb-0.7p tp:relative tp:py-0.3p tp:px-o tp:ml-0"):
    area.append(i.text)
    
area
```

```python
price=[]
for i in webpage.find_all('div',class_="flex flex-col w-33pe items-center bo tp:w-half po:w-full border-r-0"):
    price.append(i.text)
    
price
```
