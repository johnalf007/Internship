```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://en.wikipedia.org/wiki/Main_Page')
page
```

```python
webpage=BeautifulSoup(page.content)
webpage
```

```python
header_tags=[]
for i in webpage.find_all('h2',class_="mp-h2"):
    header_tags.append(i.text.split())
    
header_tags
```
