```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://meesho.com/bags-ladies/pl/p7vbp')
page
```

```python
webpage=BeautifulSoup(page.content)
webpage
```

```python
product_name=[]
for i in webpage.find_all('p',class_="Text__StyledText-sc-oo0kvp-0 bWSOET NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS NewProductCard__ProductTitle_Desktop-sc-j0e7tu-4 cQhePS"):
    product_name.append(i.text)
    
product_name
```

```python
price=[]
for i in webpage.find_all('h5',class_="Text__StyledText-sc-oo0kvp-0 hiHdyy"):
    price.append(i.text)
    
price
```

```python
discounts=[]
for i in webpage.find_all('span',class_="Text__StyledText-sc-oo0kvp-0 lnonyH"):
    discounts.append(i.text)
discounts
```

```python
import pandas as pd
df=pd.DataFrame({'Product Name':product_name, 'Price':price, 'Discounts':discounts})
df
```
