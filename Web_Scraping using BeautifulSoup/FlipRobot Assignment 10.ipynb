```python
from bs4 import BeautifulSoup
import requests
```

```python
page=requests.get('https://www.bewakoof.com/men-printed-tshirts?sort=low')
page
```

```python
webpage=BeautifulSoup(page.content)
webpage
```

```python
product_name=[]
for i in webpage.find_all('div',class_="productCardDetail"):
    product_name.append(i.text)
    
product_name
```

```python
price=[]
for i in webpage.find_all('span',class_="discountedPriceText"):
    price.append(i.text)
    
price
```

```python
image_URL=[]
for i in webpage.find_all('img',class_="productImgTag"):
    image_URL.append(i['src'])
    
image_URL
```
